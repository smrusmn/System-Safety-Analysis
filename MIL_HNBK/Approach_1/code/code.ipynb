```python
import pandas as pd
import matplotlib 
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
df = pd.read_csv('../data/Reliability_Prediction.csv')
df.head()
```

```python
df['Reliability'] = 1 / df['MTBF']
df.head()
```

```python
model = LinearRegression()
model
```

```python
dfn = df.drop(columns='MTBF')
```

```python
y = dfn['Reliability']
X = dfn.drop(columns='Reliability')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.1, shuffle=True, random_state=32)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
model.fit(X_train,y_train)
y_predicted = model.predict(X_test)
```

```python
y_predicted
```

```python
def rmse(y_test, y_predicted):
    return np.sqrt(np.mean(np.square(y_test - y_predicted)))
```

```python
rmse(y_test, y_predicted)
```

```python
x_axis = [1,2,3,4]
plt.scatter(x_axis,y_test,s=200)
plt.scatter(x_axis,y_predicted)
plt.show()
```
