```python
import pandas as pd
```

```python
df = pd.read_csv('FTA_2.csv')
df.head()
```

```python
def calculate_value(row):
    if row['Gate'] == 1:
        return row['Lower_1'] * row['Lower_2'] * row['Lower_3'] * row['Lower_4']
    else:
        return row['Lower_1'] + row['Lower_2'] + row['Lower_3'] + row['Lower_4']
```

```python
df['total_redundancy'] = df.apply(calculate_value, axis=1)
df.head()
```

```python
df.tail()
```

```python
def convert_to_exp_and_round(value):
    exp_value = "{:.1e}".format(value)
    return exp_value
```

```python
# df['Lower_1'] = df['Lower_1'].apply(convert_to_exp_and_round)
df['Lower_2'] = df['Lower_2'].apply(convert_to_exp_and_round)
df['Lower_3'] = df['Lower_3'].apply(convert_to_exp_and_round)
df['Lower_4'] = df['Lower_4'].apply(convert_to_exp_and_round)
df['total_redundancy'] = df['total_redundancy'].apply(convert_to_exp_and_round)
df.head()
```

```python
def float_check(x):
    try: 
        float(x)
        return True
    except:
        return False
```

```python
dfn = df[~df['total_redundancy'].apply(float_check)]
dfn.head()
dfn.shape
```

```python
def redundancy_recommendation(df):
    recommendations = []
    for i in range(len(df)):
        if df['total_redundancy'].iloc[i] > df['Top_Prob'].iloc[i]:
            recommendations.append('Add Redundancy')
        elif df['total_redundancy'].iloc[i] < df['Top_Prob'].iloc[i]:
            recommendations.append('Over Redundant')
        elif df['total_redundancy'].iloc[i] == df['Top_Prob'].iloc[i]:
            recommendations.append('Perfect')
        else:
            recommendations.append('No Recommendation')
    df['Recommendation'] = recommendations
    return df

```

```python
df['total_redundancy'] = df['total_redundancy'].astype(float)
df['Top_Prob'] = df['Top_Prob'].astype(float)
```

```python
redundancy_recommendation(df)
```
